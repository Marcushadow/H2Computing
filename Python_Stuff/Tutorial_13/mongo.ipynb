## Updating data

```python
import pymongo
client = pymongo.MongoClient("127.0.0.1", 27017)
db = client.get_database("entertainment")
coll = db.get_collection("movies")


result = coll.find()
print("All documents in movies collection:")
for document in result:
    print(document)

search = {'year':{'$gt':2016}}
update = {'$set':{'year':2015}}

coll.update_one(search, update) 

result = coll.find()
print("All documents in movies collection after update one:")
for document in result:
    print(document)

coll.update_many(search, update)
result = coll.find()
print("All documents in movies collection after updating all:")

for document in result:
    print(document)


search = {'year':{'$eq':201}}
update = {'$unset':{'year':0}}
coll.update_many(search, update)
result = coll.find()

print("All documents in movies collection after unset:")
for document in result:
    print(document)
client.close()
```

```python
import pymongo
client = pymongo.MongoClient("127.0.0.1", 27017)
db = client.get_database("entertainment")
coll = db.get_collection("tv")
coll.insert_one({"title":"X Man", "genre":"science fiction"})
coll.insert_one({"title":"Fresh from the boat","genre":"comedy"})
coll.insert_one({"title":"", "genre":"comedy"})
coll.insert_one({"genre":"comedy"})
result = coll.find()
print("All documents in tv collection:")
for document in result:
    print(document)
print("Number of items in tv collection:", coll.count())
db.drop_collection("tv")
result = coll.find()
print("After tv collection is dropped:")
for document in result:
    print(document)
print("Number of items in tv collection:", coll.count())
client.close()
```

```python
client = pymongo.MongoClient("127.0.0.1", 27017)

db = client.get_database("theatre")
coll = db.get_collection("concert")

def printConcert(document):
    # print(document)
    for key, value in document.items():
        if(key != "_id"):
            print(f"{key}: {value}")

    print()

menu = f"""
    Welcome to the database user interface!

    1. Add Concert (title, date, time, venue, price of tickets)
    2. Search by Title
    3. Delete by Title
    4. Update by Title
    5. Find by budget
    6. Find by month
    7. Upcoming
    8. Delete highest price
    9. Exit menu 

"""

print(menu)
answer = input("Input task: ")

while(answer != "9"):
    try: 
        answer = int(answer)
    except:
        answer = input("Input task: ")
    
    if(answer > 9):
        answer = input("Input task: ")
    
    if(answer == 1):
        reply = {}
        query = {}
        reply['title'] = input("Input title: ")
        reply['month'] = input("Input month: ")
        reply['date'] = input("Input date: ")
        reply['venue'] = input("Input venue: ")
        reply['lowestPrice'] = input("Input lowest price of tickets:")
        reply['highestPrice'] = input("Input highest price of tickets:")

        for key, value in reply.items():
            # print(key, value)
            if value != "":
                if(value.isnumeric()):
                    value = int(value)
                query[key] = value

        coll.insert_one(query)
        answer = input("Input task: ")
    
    elif(answer == 3):
        name = input("Input delete title: ")
        query = {'title': name}

        coll.delete_one(query)
        answer = input("Input task: ")
    
    elif(answer == 2):
        name = input("Input display title: ")
        query = {'title': name}

        for item in coll.find(query):
           printConcert(item)

        answer = input("Input task: ")

    elif(answer == 4):
        name = input("Input updated title: ")
        reply = {}
        query = {}
        reply['title'] = input("Input title: ")
        reply['month'] = input("Input month: ")
        reply['date'] = input("Input date: ")
        reply['venue'] = input("Input venue: ")
        reply['lowestPrice'] = input("Input lowest price of tickets:")
        reply['highestPrice'] = input("Input highest price of tickets:")

        for key, value in reply.items():
            # print(key, value)
            if value != "":
                query[key] = value
        
        condition = {"title": name}
        update = {"$set": query}
        coll.update_many(condition, update)
        answer = input("Input task: ")
    
    elif (answer == 5):
        budget = int(input("Enter your budget: "))

        query = {"lowestPrice": {"$lte": budget}}
        for item in coll.find(query):
            printConcert(item)
        
        answer = input("Input task: ")

    
    elif (answer == 6):
        month = input("Enter your month: ")

        query = {"month": month}
        for item in coll.find(query):
            printConcert(item)
        answer = input("Input task: ")

    elif (answer == 7):
        query = {"month":{"$gt":4}, "date" : {"$gt": 12}}
        printConcert(coll.find_one(query))

        answer = input("Input task: ")

    elif (answer == 8):
        query = {"$unset": {"highestPrice":0 }}

        coll.update_many({}, query)
        answer = input("Input task: ")



        

```
