````python
# Task 4.1

# Implementation of the Stack ADT using Python list.

class ListStack:
    def __init__(self):
        self.arr = []
        self.top = -1
    
    def size(self):
        return self.top + 1
    
    def pop(self):
        if(self.size() == 0):
            print("This stack is empty")
            return None
        else:
            value = self.arr[self.top]
            self.arr.pop()
            self.top -= 1
            return value
    
    def push(self, item):
        self.top += 1
        self.arr.append(item)
        
    """ List-based stack implementation """
    
   








````

````python
#  Task 4.1  

# Implementation of the Stack ADT using linked list

class Node:
    def __init__(self, data, next):
        self.data = data
        self.next = next

class LinkedStack:
    
    """ Link-based stack implementation """
    def __init__(self):
        self.top = None
        self._size = 0
    
    def size(self):
        return self._size
    
    def push(self, item):
        self._size += 1
        tempNode = Node(item, self.top)
        self.top = tempNode
        
    def pop(self):
        if(self._size == 0):
            print("The stack is empty")
            return None
    
        else:
            self._size -= 1
            value = self.top.data
            self.top = self.top.next
            return value
    
   






````
