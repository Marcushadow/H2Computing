```python
# Task 4.1

class ListNode:
    def __init__(self, value = "EMPTY", next = -1):
        self.DataValue = value
        self.PointerValue = next

    def setNext(self, newNext):
        self.PointerValue = newNext

    def getNext(self):
        return self.PointerValue

    def setValue(self, newValue):
        self.DataValue = newValue

    def getValue(self):
        return self.DataValue

class LinkedList:
    def Initialise(self):
        self.NextFree = 0
        self.Start = -1

        self.Node = []

        for i in range(10):
            temp = ListNode(next = (i+2)%11 - 1)
            self.Node.append(temp)

    def Display(self):

        print(f"start = {self.Start + 1} \t Next Free = {self.NextFree + 1}")
        for i in range(10):
            print(f"index = {i+1} \t value = {self.Node[i].getValue()} \t Pointer Value = {self.Node[i].getNext()+1}")


    def IsFull(self):
        return self.NextFree == -1
    
    def IsEmpty(self):
        return self.Start == -1

    def AddNode(self, value):
        if (self.IsFull()):
            print("It is full!")
        else:
            nextNextFree = self.Node[self.NextFree].getNext()
            

            probe = self.Start
            while(probe != -1 and self.Node[probe].getValue() < value):
                #  Probe is the bigger value
                # Probe is -1
                prev = probe
                probe = self.Node[probe].getNext()

            if(probe == self.Start):
                self.Node[self.NextFree].setValue(value)
                self.Node[self.NextFree].setNext(self.Start)
                self.Start = self.NextFree
                self.NextFree = nextNextFree

            else:

                #prev is one before
                self.Node[self.NextFree].setValue(value)
                self.Node[self.NextFree].setNext(probe)
                self.Node[prev].setNext(self.NextFree)
                self.NextFree = nextNextFree

    def RemoveNode(self, value):
        if(self.IsEmpty()):
            print("It is Empty!")

        else:
            probe = self.Start

            while(probe != -1 and self.Node[probe].getValue() != value):
                prev = probe
                probe = self.Node[probe].getNext()

            if(probe == self.Start):
                prevNextFree = self.NextFree
                prevStart = self.Start
                self.NextFree = prevStart   
                self.Node[prevStart].setValue("EMPTY") 
                self.Start = self.Node[prevStart].getNext()
                self.Node[prevStart].setNext(prevNextFree)  
                

            elif(probe == -1):
                print("No such item!")

            else:
                prevNextFree = self.NextFree
                self.Node[probe].setValue("EMPTY")
                self.Node[prev].setNext(self.Node[probe].getNext())

                self.Node[probe].setNext(self.NextFree)
                self.NextFree = probe
                

                
                      
           
            





```
