```python
# Own array implementation

# Supposed to be in alphabetical order

class DataNode:
    def __init__(self, Value, Next = 0):
        self.Value = Value
        self.Next = Next

    def IsSmaller(self, otherNode):
        if(otherNode.Value == None):
            return True
        else:
            return (self.Value <= otherNode.Value)

class LinkedListArray:
    def __init__(self, MaxSize):
        self.MaxSize = MaxSize
        self.Size = 0 
        self.NextFree = 0
        self.EOL = MaxSize + 1
        self.Start = self.EOL
        self.arr = [None] * MaxSize

        for i in range(MaxSize):
            insertNode = DataNode(None, (i+1))
            self.arr[i] = insertNode
    
    def transparentDisplay(self):
        print(f"Maximum Size\t\t: {self.Size} \nCurrent Filled Size\t: {self.Size}\nCurrent Start\t\t: {self.Start}\nCurrent Next Free\t: {self.NextFree}\nCurrent End Of List\t: {self.EOL}")
        print("Current Array State:")
        for i in range(self.MaxSize):
            print(f"Index: {i} \t\t Value:\t{self.arr[i].Value}\t\tNext: {self.arr[i].Next}")
        
    def __len__(self):
        return self.Size

    def addNode(self, addedValue):
        nextFree = self.arr[self.NextFree].Next
        self.arr[self.NextFree] = DataNode(addedValue)
        

        if(self.Size == 0):
            self.Size += 1
            self.Start = self.NextFree
            self.arr[self.NextFree].Next = self.EOL
            self.NextFree = nextFree

        else:
            if(self.Size != self.MaxSize):
                self.Size += 1

                probe = self.Start
                while(self.arr[probe].Next != self.EOL and (not self.arr[self.NextFree].IsSmaller(self.arr[self.arr[probe].Next])) ):

                    probe = self.arr[probe].Next

                if (probe == self.Start):
                    # print(f"Smallest is {addedValue} !")
                    self.arr[self.NextFree].Next = self.Start
                    self.Start = self.NextFree
                    self.NextFree = nextFree
                
                elif (self.arr[probe].Next == self.EOL):
                    self.arr[probe].Next = self.NextFree
                    self.arr[self.NextFree].Next = self.EOL
                    self.NextFree = nextFree
                
                else:
                    self.arr[self.NextFree].Next = self.arr[probe].Next
                    self.arr[probe].Next = self.NextFree
                    self.NextFree = nextFree
    
    def removeNode(self, value):
        prev = None
        probe = self.Start
        while (self.arr[probe].Next != self.EOL and self.arr[probe].Value != value):
            prev = probe
            probe = self.arr[probe].Next

        if(probe == self.EOL):
            print("Not in List")
        else:
            self.Size -= 1

            # print(self.arr[probe].Value)
            self.arr[probe].Value = None
            
            if(probe == self.Start):
                newStart = self.arr[self.Start].Next
                self.arr[self.Start].Next = self.NextFree
                self.NextFree = self.Start
                self.Start = newStart
            
            else:
                # The next of the deleted node
                newNext = self.arr[self.arr[prev].Next].Next
                self.arr[self.arr[prev].Next].Next = self.NextFree
                self.NextFree = self.arr[prev].Next

                self.arr[prev].Next = newNext

    def InOrderTraversalFromNode(self, index):
        if(index != self.EOL):
            print(self.arr[index].Value)
            self.InOrderTraversalFromNode(self.arr[index].Next)
    
    def InOrderTraversal(self):
        self.InOrderTraversalFromNode(self.Start)

    def ReverseOrderTraversalFromNode(self, index):
        if(index != self.EOL):
            self.ReverseOrderTraversalFromNode(self.arr[index].Next)
            print(self.arr[index].Value)
    
    def ReverseOrderTraversal(self):
        self.ReverseOrderTraversalFromNode(self.Start)
    



            
```
