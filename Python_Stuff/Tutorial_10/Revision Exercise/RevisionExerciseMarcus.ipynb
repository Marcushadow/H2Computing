```python
# Task 3.1 - Task 3.3
class Node:
    def __init__(self, value, next=None):
        self.value = value
        self.next = next

class DataStructure:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def display(self):
        probe = self.head
        while(probe != None):
            print(probe.value, end=" ")
            probe = probe.next
        print()
            
class Stack(DataStructure):
    def __init__(self):
        super().__init__()
    
    def insert(self, value):
        newNode = Node(value)
        newNode.next = self.head
        self.head = newNode
        
        if(self.head == None):
            self.tail = newNode
            
    
    def delete(self):
        if(self.head != None):
            value = self.head.value

            self.head = self.head.next
            return value
        else:
            return None
    
    
    
        
class Queue(DataStructure):
    def __init__(self):
        super().__init__()
    
    def insert(self, value):
        newNode = Node(value)
        if(self.head == None):
            self.head = newNode
            self.tail = newNode
            
        else:
            self.tail.next = newNode
            self.tail = newNode
    
    def delete(self):
        if(self.head != None):
            value = self.head.value
            self.head = self.head.next
            return value
        else:
            return None
    
            
    

```

```python
#  Task 3.3 - Task 3.4
class PassengerVehicle:
    def __init__(self, regNo, seatCount, vehicleType):
        self.regNo = regNo
        self.seatCount = seatCount
        self.vehicleType = vehicleType
        self.maxStanding = 0
        self.seatBeltsFitted = False
        
    def output(self):
        return [str(self.regNo), str(self.seatCount), str(self.vehicleType), str(self.maxStanding), str(self.seatBeltsFitted)]

    
class Bus(PassengerVehicle):
    def __init__(self, regNo, seatCount, vehicleType, maxStanding):
        super().__init__(regNo, seatCount, vehicleType)
        self.maxStanding = maxStanding

class Coach(PassengerVehicle):
    def __init__(self, regNo, seatCount, vehicleType, SeatBeltsFitted):
        super().__init__(regNo, seatCount, vehicleType)
        self.SeatBeltsFitted = SeatBeltsFitted
```
