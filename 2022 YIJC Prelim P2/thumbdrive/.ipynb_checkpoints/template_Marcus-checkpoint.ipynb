```python
# this Haversine function will be provided.
# students just need to know how to input 2 locations using the (longitude,latitude) coordinates 

from math import *

def Haversine(coord1,coord2):

    long1, lat1 = coord1
    long2, lat2 = coord2

    R = 6371000                               # radius of Earth in meters
    phi_1 = radians(lat1)
    phi_2 = radians(lat2)

    delta_phi = radians(lat2-lat1)
    delta_lambda = radians(long2-long1)

    a = sin(delta_phi/2.0)**2 + cos(phi_1) * cos(phi_2) * sin(delta_lambda/2.0)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))

    return R * c/1000.0


pt1 = [103.7423112,1.33315262]
pt2 = [103.74959,1.349034109]

Haversine(pt1, pt2)
```

```python
#Task 3.1

class Node:
    def __init__(self, NewTreeItem =""):
        self.data = NewTreeItem
        self.left = 0
        self.right = 0
        
    def SetLeft(self,x):
        self.left = x
    
    def SetRight(self,y):
        self.right = y
        
    def GetData(self):
        return self.data
    
    def GetLeft(self):
        return self.left
    
    def GetRight(self):
        return self.right

class Tree:
    def __init__(self):
        self.ThisTree = []
        self.Root = None
        self.NextFree = 0
        
    def AddItem(self, value):
        
        self.ThisTree.append(Node(value))
        
        if(self.Root == None):
            self.Root = self.ThisTree[self.NextFree]
        else:
            probe = self.Root
            while(probe != None):
                if(probe.GetData() > value):
                    if(probe.GetLeft() == 0):
                        probe.SetLeft(self.NextFree)
                        probe = None
                    else:
                        probe = self.ThisTree[probe.GetLeft()]

                else:
                    if(probe.GetRight() == 0):
                        probe.SetRight(self.NextFree)
                        probe = None
                    else:
                        probe = self.ThisTree[probe.GetRight()]
                        
        self.NextFree += 1
    
    def InOrderTraversal(self):
        self.InOrder(self.Root)
        
    def InOrder(self, node):
        if node.GetLeft() != 0:
            self.InOrder(self.ThisTree[node.GetLeft()])
        
        print(node.GetData(), end=" ")
        
        if(node.GetRight() != 0):
            self.InOrder(self.ThisTree[node.GetRight()])


tree = Tree()
tree.AddItem('INDIA')
tree.AddItem('NEPAL')
tree.AddItem('MALAYSIA')
tree.AddItem('SINGAPORE')
tree.AddItem('BURMA')
tree.AddItem('CANADA')
tree.AddItem('LATVIA') 
tree.InOrderTraversal()
```
