```python
class Node:
    def __init__(self, data = None, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right
        
    def getData(self):
        return self.data
    
    def setData(self, data):
        self.data = data
        
    def getLeft(self):
        return self.left
    
    def setLeft(self, left):
        self.left = left
        
    def getRight(self):
        return self.right
    
    def setRight(self, right):
        self.right = right
        
    
class Array:
    def __init__(self, n):
        self.size = n
        self.free = 0
        self.arr = []
        
        for i in range(n-1):
            self.arr.append(Node(data = None, left = i+1))
            
        self.arr.append(Node(data = None, left = None))
        
    def getFree(self):
        return self.free
    
    def setFree(self, free):
        self.free = free
        
    def display(self):
        for i in self.arr:
            print(i.left, i.data, i.right)
```

```python
class BST:
    def __init__(self, arr):
        self.tree = None
        self.arr = arr
#         arr is an instance of class Array

    def add_node(self, data):
        if(self.arr.getFree() != None):
            free = self.arr.getFree()
            self.arr.arr[free].setData(data)
            nextFree = self.arr.arr[free].getLeft()
            self.arr.arr[free].setLeft(None)
            
            probe = self.tree
            
            if(probe == None):
                self.tree =  free
            
            else:
                while(probe != None):
                    if(self.arr.arr[probe].data > data):
                        if(self.arr.arr[probe].getLeft() == None):
                            self.arr.arr[probe].setLeft(free)
                            probe = None
                        else:
                            probe = self.arr.arr[probe].getLeft()
                    else:
                        if(self.arr.arr[probe].getRight() == None):
                            self.arr.arr[probe].setRight(free)
                            probe = None
                        else:
                            probe = self.arr.arr[probe].getRight()
                            
            self.arr.setFree(nextFree)
        else:
            print("Is full")
                            
    def in_order_traversal(self):
        self.in_order(self.tree)
       
    def in_order(self, node):
        if(node != None):
            if(self.arr.arr[node].getLeft() != None):
                self.in_order(self.arr.arr[node].getLeft())
            print(self.arr.arr[node].data, end=" ")
            if(self.arr.arr[node].getRight() != None):
                self.in_order(self.arr.arr[node].getRight())
                
    def search_tree(self, data):
        if(self.tree == None):
            return "Not inside"
        
        else:
            position = "Not inside"
            probe = self.tree
            while(probe != None):
                if(self.arr.arr[probe].getData() == data):
                    position = probe
                    probe = None
                
                elif(self.arr.arr[probe].getData() > data):
                    probe = self.arr.arr[probe].getLeft()
                    
                else:
                    probe = self.arr.arr[probe].getRight()
        
        return position
```

```python
arr = Array(10)

test = BST(arr)
test.add_node(1)
test.add_node(2)

arr.display()

test.in_order_traversal()

test.search_tree(3)
```
