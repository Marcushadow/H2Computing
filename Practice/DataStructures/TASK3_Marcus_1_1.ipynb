```python
class Node:
    def __init__(self, nxt):
        self.Data = ""
        self.Priority = 0
        self.Next = nxt
        
class PQueue:
    def Initialise(self):
        self.ThisPQueue = [None]*11
        self.Front = 0
        self.NextFree = 1
        self.Rear = 0
        
        
        for i in range(1, 10):
            self.ThisPQueue[i] = Node(i+1)
            
        self.ThisPQueue[10] = Node(0)
        
        
    def PQInsert(self, NewItem, Priority):
        newNode = Node(0)
        newNode.Priority = Priority
        newNode.Data = NewItem
        nextnextFree = self.ThisPQueue[self.NextFree].Next
        self.ThisPQueue[self.NextFree] = newNode
        
        if(self.Front == 0):
            self.Front = self.NextFree
            
        else:
            self.ThisPQueue[self.Rear].Next = self.NextFree
        
        
        
        self.Rear = self.NextFree
        self.NextFree = nextnextFree
    
    def PQDelete(self):
        prev = 0
        probe = self.Front
        currPrev = 0        
        hiPri = self.ThisPQueue[self.Front].Priority
        
        while(probe != 0):
            if(self.ThisPQueue[probe].Priority < hiPri):
                hiPri = self.ThisPQueue[probe].Priority
                currPrev = prev
                
            prev = probe
            probe = self.ThisPQueue[probe].Next
        
        if(currPrev == 0):
            oldHead = self.Front
            self.Front = self.ThisPQueue[self.Front].Next
            self.ThisPQueue[oldHead].Next = self.NextFree
            self.NextFree = oldHead
            
        else:
            toDelete = self.ThisPQueue[prev].Next
            self.ThisPQueue[prev].Next = self.ThisPQueue[toDelete].Next
            self.ThisPQueue[toDelete].Next = self.NextFree
            self.NextFree = toDelete
        
        value = self.ThisPQueue[self.NextFree].Data
        self.ThisPQueue[self.NextFree].Data = ""
        self.ThisPQueue[self.NextFree].Priority = 0
        
        if(self.Front == 0):
            self.Rear = 0
        
            
        return value
            
        
    
    def DisplayPQueue(self):
        print(f"Front: {self.Front}, Rear: {self.Rear}, NextFree: {self.NextFree}")
        for i in range(1, 11):
            print(f"index: {i}, Data: {self.ThisPQueue[i].Data}, Priority: {self.ThisPQueue[i].Priority}, Next: {self.ThisPQueue[i].Next}")
```

```python
testPQ = PQueue()

testPQ.Initialise()
testPQ.PQInsert(1,1)
testPQ.PQInsert(2,3)

testPQ.PQDelete()

testPQ.DisplayPQueue()
```
