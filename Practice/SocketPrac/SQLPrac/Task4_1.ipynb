```python
import csv, sqlite3
client = sqlite3.connect("voting_mgm.db")

with open("students.csv") as fs:
    reader = csv.reader(fs, delimiter =",")
    headers = next(reader)
        
    
    for item in reader:
        curr = client.execute("INSERT INTO Student (MatricNo, `Class`, IndexNo, Gender) VALUES (?,?,?,?)", (item[0], item[1], item[2], item[3]))
    
with open("candidates.csv") as fs:
    reader = csv.reader(fs, delimiter =",")
    headers = next(reader)

    for item in reader:
        curr = client.execute("INSERT INTO Candidate (CandidateNo, Name, Slogan, PortraitLink) VALUES (?,?,?,?)", (item[0], item[1], item[2], item[3]))
    
with open("votes.csv") as fs:
    reader = csv.reader(fs, delimiter =",")
    headers = next(reader)
    
    for item in reader:
        curr = client.execute("INSERT INTO Vote (MatricNo, CandidateNo) VALUES (?,?)", (item[0], item[1]))
    
client.commit()
client.close()
```
