```python
class Node:
    def __init__(self, data = "", nxt = -1):
        self.data = data
        self.nxt = nxt

class LinkedStructure:
    def Initialise(self):
        self.Head = -1
        self.Tail = -1
        self.Free = 0
        self.LinkedList = []
        
        for i in range(10):
            tempNode = Node("", ((i+2)%11)-1)
            self.LinkedList.append(tempNode)
            
    def IsEmpty(self):
        return self.Head == -1

    def IsFull(self):
        return self.Free == -1
    
    def Display(self):
        probe = self.Head
        
        while(probe != -1):
            print(self.LinkedList[probe].data)
            probe = self.LinkedList[probe].nxt
    
    def PrintStructure(self):
        print("Structure: ")
        print(f"Head: {self.Head}, Tail: {self.Tail}, Free: {self.Free}")
        for i in range(len(self.LinkedList)):
            item = self.LinkedList[i]
            print(f"Index: {i} , Data: {item.data}, Nxt: {item.nxt}")
            
    def Add(self, item):
        if self.IsFull():
            print("Error: Full")
            
        else:
            nextFree = self.LinkedList[self.Free].nxt
            self.LinkedList[self.Free].data = item
            prev = -1
            probe = self.Head
            
            while (probe != -1 and self.LinkedList[probe].data < item):
                prev = probe
                probe = self.LinkedList[probe].nxt
            
            self.LinkedList[prev].nxt = self.Free
            self.LinkedList[self.Free].nxt = probe
            
            if(prev == -1):
                self.Head = self.Free
            
            if(probe == -1):
                self.Tail = self.Free
            
            
            self.Free = nextFree
                        
    def Remove(self, item):
        if self.IsEmpty():
            print("Error is Empty")
        else:
            prev = -1
            probe = self.Head
            
            while(probe != -1 and self.LinkedList[probe].data != item):
                prev = probe
                probe = self.LinkedList[probe].nxt
                
            if(probe == -1):
                print("Error: No exist")
            
            else:
                if(prev == -1):
                    self.Head = self.LinkedList[probe].nxt
                else:
                    self.LinkedList[prev].nxt = self.LinkedList[probe].nxt
                
                self.LinkedList[probe].nxt = self.Free
                self.LinkedList[probe].data = ""
                self.Free = probe
                
                if(self.Tail == probe):
                    self.Tail = prev
                
            
    
        
```

```python
# Task 4.2
Country = LinkedStructure()
Country.Initialise()

Country.Add("Japan")
Country.Add("Singapore")
Country.Add("China")

print()
Country.PrintStructure()

print()
print("Display: ")
Country.Display()

Country.Remove("China")
Country.Remove("Japan")

print()
Country.PrintStructure()
```

```python
class Queue(LinkedStructure):
    def __init__(self):
        super().Initialise()
        
    def Add(self, item):
        if(self.IsFull()):
            print("Error: Full")
            
        else:
            nextFree = self.LinkedList[self.Free].nxt
            self.LinkedList[self.Free].data = item

            if(self.Head == -1):
                self.Head = self.Free
                
            
            else:
                self.LinkedList[self.Tail].nxt = self.Free
                self.LinkedList[self.Free].nxt = -1
            
            self.Tail = self.Free
            self.Free = nextFree
    def Remove(self):
        if(self.IsEmpty()):
            print("Error: Is empty")
        
        else:
            print(self.LinkedList[self.Head].data)
            
            oldHead = self.Head
            self.Head = self.LinkedList[self.Head].nxt
            self.LinkedList[oldHead].nxt = self.Free
            self.Free = oldHead
            self.LinkedList[self.Free].data = ""
            
            if(self.Head == -1):
                self.Tail = -1

    def Display(self):
        
        print("\nDisplaying: ")
        probe = self.Head
        
        while(probe != self.Tail):
            print(self.LinkedList[probe].data)
            probe = self.LinkedList[probe].nxt
        print(self.LinkedList[probe].data)
        
    
            
                
            
            
```
