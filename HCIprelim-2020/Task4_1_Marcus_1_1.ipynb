```python
#Task 4.1

class Node:
    def __init__(self, data ="", iNext = 0):
        self.Data = data
        self.Next = iNext
    
    def setData(self, data):
        self.Data = data
    
    def getData(self):
        return self.Data
    
    def setNext(self, iNext):
        self.Next = iNext
    
    def getNext(self):
        return self.Next

class LinkedStructure:
    def Initialise(self):
        self.LLA = []

        for i in range(11):
            temp = Node()
            temp.setNext((i+1)%11 )
            self.LLA.append(temp)

        self.Head = 0
        self.Tail = 0
        self.Free = 1
    
    def PrintStructure(self):
        print(f"Head: {self.Head}, Tail: {self.Tail}, Free: {self.Free}")
        for i in range(1, 11):
            print(f"Index: {i}, Data: {self.LLA[i].getData()}, Next: {self.LLA[i].getNext()}")
    
    def IsFull(self):
        return self.Free == 0

    def IsEmpty(self):
        return self.Head == 0
    
    def Display(self):
        if(self.IsEmpty()):
            print("Empty")
        else:
            probe = self.Head
            while(probe != 0):
                print(self.LLA[probe].getData())
                probe = self.LLA[probe].getNext()
            
            
    
    def Add(self, item):
        if(self.IsFull()):
            print("Is full")
        else:
            self.LLA[self.Free].setData(item)
            nextFree = self.LLA[self.Free].getNext()
            self.LLA[self.Free].setNext(0)

            if(self.IsEmpty()):
                self.Head = self.Free
                self.Tail = self.Free

            else:
                prev = 0
                probe = self.Head
                while(probe != 0 and self.LLA[probe].getData() < item):
                    prev = probe
                    probe = self.LLA[probe].getNext()

                if(probe == 0):
                    self.Tail = self.Free

                if(prev == 0):
                    self.Head = self.Free
                else:
                    self.LLA[prev].setNext(self.Free)

                self.LLA[self.Free].setNext(probe)
                
            self.Free = nextFree

    def Remove(self, item):
        if(self.IsEmpty()):
            print("Failed as empty")
        else:
            prev = 0
            probe = self.Head
            while(probe != 0 and self.LLA[probe].getData() != item):
                prev = probe
                probe = self.LLA[probe].getNext()

            if(probe == self.Tail):
                self.Tail = prev

            if(probe == 0):
                print("Not found")
            else:
                self.LLA[probe].setData('')

                if(prev == 0):
                    self.Head = self.LLA[probe].getNext()
                else:
                    self.LLA[prev].setNext(self.LLA[probe].getNext())

                self.LLA[probe].setNext(self.Free)
                self.Free = probe







            
```

```python
#Task 4.2

country = LinkedStructure()
country.Initialise()

country.Add("Japan")
country.Add("Singapore")
country.Add("China")

country.PrintStructure()
print()
country.Display()
print()

country.Remove("China")
country.Remove("Japan")
country.PrintStructure()
```
