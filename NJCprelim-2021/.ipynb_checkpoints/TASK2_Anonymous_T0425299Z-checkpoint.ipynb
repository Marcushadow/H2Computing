```python
# Task 2

class Employee:
    def constructor(self, name, NRIC, contact):
        self.name = name
        self.NRIC = NRIC
        self.contact = contact
        
    def __repr__(self):
        return f"{self.NRIC} : {self.name}"
    
    def compare_with(self, object):
        if(self.name < object.name):
            return -1
    
        elif(self.name > object.name):
            return 1
    
        else:
            return 0
    
        
```

```python
#Task 2.2

class FTEmployee(Employee):
    def constructor(self, name, NRIC, contact, monthlySalary):
        super().constructor(name, NRIC, contact)
        self.monthlySalary = monthlySalary
    
    def getMonthlySalary(self):
        return self.monthlySalary
    
    def setMonthlySalary(self, newSalary):
        self.monthlySalary = newSalary
    
    def __repr__(self):
        return super().__repr__() + f": {self.calculate_pay()}"
    
    def calculate_pay(self):
        return self.getMonthlySalary()

    def compare_with(self, object):
        if(int(self.calculate_pay()) < int(object.calculate_pay())):
            return -1
    
        elif(int(self.calculate_pay()) > int(object.calculate_pay())):
            return 1
    
        else:
            return super().compare_with(object)


        
class PTEmployee(Employee):
    def constructor(self, name, NRIC, contact, hourlyRate, daysPerMonth):
        super().constructor(name, NRIC, contact)
        self.hourlyRate = hourlyRate
        self.daysPerMonth = daysPerMonth
    
    def getDaysPerMonth(self):
        return self.daysPerMonth
    
    def setDaysPerMonth(self, daysPerMonth):
        self.daysPerMonth = daysPerMonth
        
    def getHourlyRate(self):
        return self.hourlyRate
    
    def setHourlyRate(self, hourlyRate):
        self.hourlyRate = hourlyRate
    
    def __repr__(self):
        return super().__repr__() + f": {self.calculate_pay()}"
    
    def calculate_pay(self):
        return int(self.getDaysPerMonth()) * int(self.getHourlyRate())

    def compare_with(self, object):
        if(int(self.calculate_pay()) < int(object.calculate_pay())):
            return -1
    
        elif(int(self.calculate_pay()) > int(object.calculate_pay())):
            return 1
    
        else:
            return super().compare_with(object)


```

```python
import csv

with open("PAID_EMPLOYEES.CSV") as fs:
    reader = csv.reader(fs, delimiter=",")
    headRow = next(reader)
    paidEmployee = []
    
    for row in reader:
    
        name = row[2]
        NRIC = row[1]
        contact = row[3]
        hourlyRate = row[4]
        daysPerMonth = row[5]
        monthlySalary = row[6]
    
        if(row[0] == "PT"):
            temp = PTEmployee()
            temp.constructor(name, NRIC, contact, hourlyRate, daysPerMonth)
            paidEmployee.append(temp)

        else:
            temp = FTEmployee()
            temp.constructor(name, NRIC, contact, monthlySalary)
            paidEmployee.append(temp)
        
        
    print(paidEmployee)
```

```python
#Task 2.4

tempPaidEmployee = paidEmployee
def insert_sort(paidEmployeeList):
    for i in range(1, len(paidEmployeeList)):
        j = i
        while(paidEmployeeList[j].compare_with(paidEmployeeList[j-1]) == -1 and (j > 0)):
            paidEmployeeList[j], paidEmployeeList[j-1] = paidEmployeeList[j-1], paidEmployeeList[j]
            j = j-1
            
insert_sort(tempPaidEmployee)
print(tempPaidEmployee)
    
    
```

```python
# Task 2.5

import csv

with open("ALL_EMPLOYEES.CSV") as fs:
    reader = csv.reader(fs, delimiter=",")
    headRow = next(reader)
    
    startOfPaid = 0
    all_employee = []
    
    for row in reader:
        employeeType = row[0]
        name = row[2]
        NRIC = row[1]
        contact = row[3]
        hourlyRate = row[4]
        daysPerMonth = row[5]
        monthlySalary = row[6]
        
        if(employeeType == "PT"):
            temp = PTEmployee()
            temp.constructor(name, NRIC, contact, hourlyRate, daysPerMonth)
            all_employee.append(temp)
        
        elif(employeeType == "FT"):
            temp = FTEmployee()
            temp.constructor(name, NRIC, contact, monthlySalary)
            all_employee.append(temp)
        
        else:
            temp = Employee()
            temp.constructor(name, NRIC, contact)
            all_employee.insert(0, temp)
            startOfPaid += 1
```

```python

def new_insert_sort(paidEmployeeList, start, end):

    for i in range(start+1, end+1):

        j = i
        while((j > start) and paidEmployeeList[j].compare_with(paidEmployeeList[j-1]) == -1):
            paidEmployeeList[j], paidEmployeeList[j-1] = paidEmployeeList[j-1], paidEmployeeList[j]
            j = j-1


new_insert_sort(all_employee, 0, startOfPaid-1)
new_insert_sort(all_employee, startOfPaid, len(all_employee)-1)
print(all_employee)
```
