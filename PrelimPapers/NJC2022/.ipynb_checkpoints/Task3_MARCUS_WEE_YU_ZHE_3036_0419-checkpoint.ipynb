```python
# Task 3.1

class Page:
    def __init__(self, addr="", title="", stamp=""):
        self.page_address = addr
        self.page_title = title
        self.time_stamp = stamp
        
    def get_address(self):
        return self.page_address
    
    def get_title(self):
        return self.page_title
    
    def get_date(self):
        return self.time_stamp[:13]
    
    def get_hour(self):
        return self.time_stamp[14:]
    
    def display(self):
        return f"Page({self.get_address()}, {self.get_title()}, {self.time_stamp})"
```

```python
# Task 3.2
import csv
class Node:
    def __init__(self, data= None, ptr = None):
        self.data = data
        self.ptr = ptr

class Stack:
    def __init__(self):
        self.head = None
        
    def push(self, page_object):
        newNode = Node(data = page_object, ptr = self.head)
        self.head = newNode
        
    def pop(self):
        returnVal = self.head.data
        self.head = self.head.ptr
        
        return returnVal
    
    def search(self,page_title):
        probe = self.head
        found = False
        while(probe != None and not found):
            if(probe.data.get_title() == page_title):
                found = True
                
            probe = probe.ptr
        return found    
    
    def count(self):
        ctr = 0
        probe = self.head
        while(probe != None):
            ctr += 1
            probe = probe.ptr
        
        return ctr
    
    def to_String(self):
        outputStr = "["
        probe = self.head
        while(probe != None):
            
            outputStr += probe.data.display()
            
            if(probe.ptr != None):
                outputStr += ", "
            
            probe = probe.ptr
        
        outputStr += "]"
        return outputStr
    
    
s = Stack()
with open("Task3data1.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    header = next(reader)
    for i in reader:
        tempPage = Page(addr=i[1], title=i[2], stamp=i[3])
        s.push(tempPage)
        
print(s.to_String())
    
# test = Stack()
# test.push(Page(title = "Work"))
# # test.search("Work")
# test.to_String()
```

```python
# Task 3.3

class ModStack(Stack):
    def __init__(self):
        super().__init__()
        
    def push(self, page_object):
        if(self.count() >= 12):
            prev = None
            probe = self.head
            while(probe.ptr!= None):
                prev = probe
                probe = probe.ptr
            prev.ptr = None
        
        super().push(page_object)
        
ms = ModStack()
with open("Task3data2.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    header = next(reader)
    for i in reader:
        tempPage = Page(addr=i[1], title=i[2], stamp=i[3])
        ms.push(tempPage)
        
print(ms.to_String())
```
