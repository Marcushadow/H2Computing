```python
import pymongo, csv

client = pymongo.MongoClient("127.0.0.1", 27017)
db = client.get_database("GovTech")
coll = db.get_collection("Datasets")


with open("data-gov-sg-dataset-listing.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    headings = next(reader)
    
    for row in reader:
        tempDict = {}
        
        newRow = row
        newRow[0] = int(newRow[0])
        
        for i in range(len(newRow)):
            if(newRow[i] != "na"):
                tempDict[headings[i]] = newRow[i]
            
        coll.insert_one(tempDict)
            
        
```
