```python
class Character:
    def __init__(self, name = "", hp = 0, mp = 0, atk = 0):
        self.name = ""
        self.hp = 0
        self.mp = 0
        self.atk = 0
        
    def get_name(self):
        return self.name
    
    def get_hp(self):
        return self.hp
    
    def get_mp(self):
        return self.mp

    def get_atk(self):
        return self.atk
    
    def update_hp(self, hp_change):
        self.hp += hp_change
        
    def update_mp(self, mp_change):
        self.mp += mp_change
    
    def __str__(self):
        return f"name: {self.name}, hp: {self.hp}, mp: {self.mp}, atk: {self.atk}."

class Pokemon(Character):
    def __init__(self, name = "", hp = 0, mp = 0, atk = 0, p_type = ""):
        super().__init__(name, hp, mp, atk)
        self.p_type = p_type
        
    def get_type(self):
        return self.p_type

    def __str__(self):
        return f"name: {self.name}, hp: {self.hp}, mp: {self.mp}, atk: {self.atk}, type: {self.p_type}."

class Player(Character):
    def __init__(self, name = "", hp = 0, mp = 0, atk = 0):
        super().__init__(name, hp, mp, atk)
        self.pokemons = []
    
    def display_pokemons(self):
        if(len(self.pokemons) == 0):
            print("You have not caught any pokemons yet.")
        else:
            print("Here are the pokemons in your team:")
            for pokemon in self.pokemons:
                print(pokemon)
                
    def catch_pokemon(self, new_pokemon):
        replace = 0
        
        while(replace < len(self.pokemons)):
            if(self.pokemons[replace].get_type() == new_pokemon.get_type()):
                print("You already have: ")
                print(self.pokemons[replace])
                print("Now you meet:")
                print(new_pokemon)
                
                doReplace = input("Would you like to replace the pokemon of the same type? [Y/N]")
                
                if(doReplace == "Y"):
                    print(f"You have released {self.pokemons[replace].get_name()}, and caught {new_pokemon.get_name()}.")
                    self.pokemons[replace] = new_pokemon
                else:
                    print(f"You chose not to catch {new_pokemon.get_name()}.")
                replace = len(self.pokemons)+1
                
            else:
                replace += 1
            
            
        if(replace == len(self.pokemons)):
#             Did not make a replacement
            self.pokemons.append(new_pokemon)
            print(f"You have caught {new_pokemon.get_name()}.")
```
