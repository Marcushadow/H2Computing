```python
# Task 2.1

def validate_guesscode(guesscode):
    guesscode = str(guesscode)
    valid = True
    if(len(guesscode) != 4):
        valid = False
    elif(not guesscode.isnumeric()):
        valid = False
    else:
        for char in guesscode:
            if (int(char)>6 or int(char)<1):
                valid = False
                
    return valid

print(validate_guesscode("6551"))
```

```python
import socket

def menu():
    gameMenu = """
    Master Mind Game Menu
    1) Just start a new game
    2) Make a guess code
    3) Display history
    4) Quit
    """
    print("Client initiates connection to server")
    clientSock = socket.socket()
    clientSock.connect(("127.0.0.1", 12345))
    print("Client connected to server")
    
    print(gameMenu)
    choice = input("Type an option:")
    
    while(choice != "4"):
        if(choice == "1"):
            clientSock.sendall(b"START\n")
            message = clientSock.recv(1024).decode()
            attempt = message.strip("\n").split(",")[1]
            print(f"New game starts begins with {attempt} attempts.")
            
        elif(choice == "2"):
            guess = input("Make a guess:")
            guessMSG = "GUESS,"+ guess+"\n"
            clientSock.sendall(guessMSG.encode())
            
            message = clientSock.recv(1024).decode().strip("\n").split(",")
            if(message[0] == "WON"):
                print(f"You have WON. The code is {guess}.")
            elif(message[0] == "LOSE"):
                print(f"You have run out of attemps. The code is {message[1]}")
            else:
                print(f"Guess Result: {message[1]}")
                
        elif(choice == "3"):
            clientSock.sendall(b"HISTORY\n")
            message = clientSock.recv(2048).decode()
            data = message.strip("\n").split(",")
            print(f"You have {data[1]} attempts left.")
            
            for i in range(2, len(data), 2):
                print(f"{i//2} {data[i]} {data[i+1]}")
        else:
            clientSock.sendall(b"QUIT\n")
            print("Game program ends.")
            
        if(choice in ["1", "2", "3"]):
            print(gameMenu)
            choice = input("Type an option:")
        else:
            choice = "4"
        
menu()            
                
            
            
    
    
```
