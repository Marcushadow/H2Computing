```python
# Task 1.1
class TreeNode:
    def __init__(self, data = ""):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None        
        
    def insert(self, value):
        newNode = TreeNode(data = value)
        probe = self.root
        if(self.root == None):
            self.root = newNode
        else:
            while(probe != None):
                if(value <= probe.data):
                    if(probe.left == None):
                        probe.left = newNode
                        probe = None
                    else:
                        probe = probe.left
                else:
                    if(probe.right == None):
                        probe.right = newNode
                        probe = None
                    else:
                        probe = probe.right
                    
    def inOrder(self, node):
        if(node != None):
            return self.inOrder(node.left) + node.data + self.inOrder(node.right)
        else:
            return ""
                    
    def inOrderTraversal(self):
        return self.inOrder(self.root)
        
    def preOrder(self, node):
        if(node != None):
            return node.data + self.preOrder(node.left) + self.preOrder(node.right)
        else:
            return ""
                    
    def preOrderTraversal(self):
        return self.preOrder(self.root)

    def postOrder(self, node):
        if(node != None):
            return self.postOrder(node.left) + self.postOrder(node.right) + node.data
        else:
            return ""
    def postOrderTraversal(self):
        return self.postOrder(self.root)
        
    def countLeaves(self, node):
        if(node != None):
            if(node.left == node.right == None):
                return 1
            else:
                return self.countLeaves(node.left) + self.countLeaves(node.right)
        else:
            return 0
        
    def leafCount(self):
        return self.countLeaves(self.root)

def task1_1():
    with open("Task1_1.txt") as fs:
        lines = fs.read().split("\n")[:-1]
        for i in range(len(lines)):
            tree = BST()
            for letter in lines[i]:
                tree.insert(letter)    
            
            print(f"{i+1} {tree.leafCount()} {tree.preOrderTraversal()} {tree.inOrderTraversal()} {tree.postOrderTraversal()}")
        
task1_1()
```
