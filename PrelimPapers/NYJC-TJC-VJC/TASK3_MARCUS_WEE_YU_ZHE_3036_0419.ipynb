```python
# Task 3.3

class HashTable:
    def __init__(self):
        self.arr = []
        for i in range(31):
            self.arr.append('')
        
    def hashAlg(self, value):
        return (value)%31
    
    def hashAlg2(self, value):
        return (2**value)%31
    
    def add(self, value, name):
        addPos = self.hashAlg(value)
        if(self.arr[addPos] != ""):
            addPos = self.hashAlg2(value)
            if(self.arr[addPos] == ""):
                self.arr[addPos] = name
        else:
            self.arr[addPos] = name
    
    def display(self):
        for i in range(31):
            print(f"index: {i}, value: {self.arr[i]}")
            
    def search(self, membershipNo):
        membershipNo = self.hashAlg(membershipNo)
        if(self.arr[membershipNo] == ""):
            print("Not found")
        else:
            print(self.arr[membershipNo])
            
        
```

```python

ht = HashTable()

with open("IDs.txt") as fs:
    data = fs.read().split("\n")
    for kv in data:
        kv = kv.split(",")
        pos = int(kv[0])
        value = kv[1]
        
        ht.add(pos, value)
ht.display()
```
