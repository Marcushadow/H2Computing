```python
import csv, sqlite3
client = sqlite3.connect("car_rental.db")

with open("customers.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    customerHeader = next(reader)
    for row in reader:
        client.execute("INSERT INTO Customer(CustomerID, Name, Gender, Contact) VALUES (?,?,?,?)", (row[0], row[1], row[2], row[3]))
        
with open("cars.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    carHeader = next(reader)
    for row in reader:
        client.execute("INSERT INTO Car(VIN, Brand, VehicleType, EnergySource, DailyPrice, Availability) VALUES (?,?,?,?,?,?)" , (row[0], row[1], row[2], row[3], float(row[4]), row[5]))

with open("rental_points.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    rentalPointsHeader = next(reader)
    for row in reader:
        
        client.execute("INSERT INTO RentalPoint(PointID, Address, OpWeekDay, OpStartHr, OpEndHr) VALUES (?,?,?,?,?)" , (row[0], row[1], row[2], row[3], row[4]))


with open("rental_records.csv") as fs:
    reader = csv.reader(fs, delimiter=",")
    rentalRecordsHeader = next(reader)
    for row in reader:
        client.execute("INSERT INTO RentalRecord(CustomerID, VIN, StartDate, CollectionPointID, ReturnDate, ReturnPointID) VALUES (?,?,?,?,?,?)" , (row[0], row[1], row[2], row[3], row[4], row[5]))

client.commit()
client.close()
```

```python
import flask, sqlite3

from flask import render_template, request

app = flask.Flask(__name__, template_folder="./Task4_4/templates")


def validVIN(inpVal):
    valueMap = {
        "A": 1,
        "B": 2,
        "C": 3,
        "D": 4,
        "E": 5,
        "F": 6,
        "G": 7,
        "H": 8,
        "J": 1,
        "K": 2,
        "L": 3,
        "M": 4,
        "N": 5,
        "P": 7,
        "R": 9,
        "S": 2,
        "T": 3,
        "U": 4,
        "V": 5,
        "W": 6,
        "X": 7,
        "Y": 8,
        "Z": 9,
    }
    weights = [8,7,6,5,4,3,2,10,0,9,8,7,6,5,4,3,2]
    sumAll = 0
    currInd = 0
    
    for character in inpVal:
        if(character in valueMap.keys()):
            sumAll += valueMap[character] * weights[currInd]
        else:
            sumAll += int(character) * weights[currInd]
        currInd += 1
    
    remainder = sumAll % 11
    if remainder == 10:
        remainder = "X"
    
    remainder = str(remainder)
    
    if(remainder == inpVal[8]):
        return True
    else:
        return False


@app.route("/")
def main():
    return render_template("index.html")

@app.route("/viewResults", methods=["POST"])
def returnResults():

    if(request.method == "POST"):
#         print("Reached, 1M8GDM9AXKP042788")
#         print(request.form.get("VIN"))

        VIN = request.form.get("VIN")
        Brand = request.form.get("Brand")
        VehicleType = request.form.get("VehicleType")
        EnergySource = request.form.get("EnergySource")
        DailyPrice = request.form.get("DailyPrice")
        Availability = "Available"
        
        if(validVIN(VIN)):

            attribs = [VIN, Brand,VehicleType, EnergySource, DailyPrice, Availability]
#             print(attribs)
            client = sqlite3.connect("car_rental.db")
            curr = client.execute("INSERT INTO Car(VIN, Brand, VehicleType, EnergySource, DailyPrice, Availability) VALUES (?,?,?,?,?,?)", tuple(attribs))
            
        
            return render_template("registration.html", attr=attribs)
        
        else:
            return render_template("failed.html")
    

if __name__ == "__main__":
    app.run()

client.commit()

```
