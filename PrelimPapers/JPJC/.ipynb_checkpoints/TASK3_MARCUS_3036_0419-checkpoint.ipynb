```python
import json, pymongo
with open("RENTAL.JSON") as fs:
    data = fs.read()
    listOfDict = json.loads(data)
    
client = pymongo.MongoClient("127.0.0.1", 27017)
db = client.get_database("StyleTheory")
coll = db.get_collection("Rental")
    
coll.insert_many(listOfDict)

client.close()
```

```python
import json, pymongo
menu = """

Style Theory
Option 1 - Insert new rental
Option 2 - Update existing rental
Option 3 - Quit
"""
client = pymongo.MongoClient("127.0.0.1", 27017)
db = client.get_database("StyleTheory")
coll = db.get_collection("Rental")
choice = input("Enter you option (1,2, or 3): ")

while(choice != "3"):

    if(choice == "1"):

        inputDict = {}

        inputDict["catalogueNumber"] = int(input("Enter Catalogue Number: "))
        inputDict["brand"] = input("Enter Brand: ")
        inputDict["category"] = input("Enter Category: ")
        inputDict["rental"] = int(input("Enter rental: "))
        inputDict["email"] = input("Enter email: ")
        inputDict["startDate"] = input("Enter start date: ")
        inputDict["endDate"] = input("Enter end date: ")

        if(inputDict["category"] == "apparel"):
            inputDict["size"] = int(input("Enter size: "))

        coll.insert_one(inputDict)


    elif(choice == "2"):
        query = {}
        query['catalogueNumber'] = int(input("What catalogue Number: "))
        query['startDate'] = input("What Start Date: ")
        newEnd = input("What new end Date: ")
        update = {"$set": {'endDate': newEnd}}

        coll.update_one(query, update)

    else:
        choice = "3"
    
    choice = input("Enter you option (1,2, or 3): ")
    

client.close()
    
    
```

```python
from datetime import date

def display_all():
    client = pymongo.MongoClient("127.0.0.1", 27017)
    db = client.get_database("StyleTheory")
    coll = db.get_collection("Rental")
    
    output = coll.find({})
    for item in output:
        print("")
        for k,v in item.items():
            if(k != "_id"):
                print(k,":",v)
                
        if("size" not in item.keys()):
            print("size : NA")
            
        f_date = date(int(item['startDate'][:4]), int(item['startDate'][5:7]), int(item['startDate'][-2:]))
        l_date = date(int(item['endDate'][:4]), int(item['endDate'][5:7]), int(item['endDate'][-2:]) )
        delta = l_date - f_date
        
        print(f"Total Payable: {delta.days * item['rental']}")
        
    client.close()
    
display_all()
```
