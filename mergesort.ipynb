```python
def quicksort(arr, left = -1, right = -1):
    if left == right == -1:
        left = 0
        right = len(arr)-1

    if(left < right):
        pivot = partition(arr, left, right);
        quicksort(arr, left, pivot -1)
        quicksort(arr, pivot + 1, right)

def partition(arr, left, right):
    mid = (left + right) //2
    pivVal = arr[mid]
    arr[mid], arr[left] = arr[left], arr[mid]

    l = left + 1
    r = right

    while(l <= r):
        while( (l <= r) and arr[l] <= pivVal):
            l += 1
        
        while( (l<= r) and arr[r] > pivVal):
            r -= 1
        
        if(l < r):
            arr[l], arr[r] = arr[r], arr[l]

    arr[r], arr[left] = arr[left], arr[r]
    return r
```

```python
class CircularQueue:
    def __init__(self, size=10):
        self.arr = []
        self.size = size
        self.head = -1
        self.tail = -1

        for i in range(self.size):
            self.arr.append('')

    def next(self, index):
        return (index+1)%self.size

    def isEmpty(self):
        return self.head == self.tail == -1
    
    def isFull(self):
        return self.next(self.tail) == self.head

    def add(self, value):
        if(self.isEmpty()):
            self.head = 0
            self.tail = 0

            self.arr[self.head] = value

        elif(not self.isFull()):
            nextIdx = self.next(self.tail)
            self.arr[nextIdx] = value
            self.tail = nextIdx

    def delete(self):
        if(not self.isEmpty()):
            value = self.arr[self.head]
            self.arr[self.head] = ""

            if(self.head == self.tail):
                self.head = -1
                self.tail = -1
            
            else:
                self.head = self.next(self.head)

    
    def display(self):
        print(f"head : {self.head}, tail : {self.tail}")

        for i in range(self.size):
            print(i, self.arr[i])



```

```python
def quickSort(arr, left = -1, right = -1):
    if(left == right == -1):
        left = 0
        right = len(arr)-1

    if(left < right):
        piv = partition(arr, left, right)
        quickSort(arr, left, piv-1)
        quickSort(arr, piv+1, right)

def partition(arr, left, right):
    mid = (left + right)//2
    arr[left], arr[mid] = arr[mid], arr[left]
    pivVal = arr[left]

    l = left +1
    r = right
    while(l <= r):
        while( (l<=r) and arr[l] <= pivVal):
            l+= 1
        
        while(arr[r] > pivVal):
            r-=1

        if(l< r):
            arr[l], arr[r] = arr[r], arr[l]

    arr[left], arr[r] = arr[r], arr[left]
    return r

    
```

```python
def convertToBase(num, base):
    output = ""
    hexString = "0123456789ABCDEF"
    while num != 0:
        remainder = num % base
        output = hexString[remainder] + output

        num = num//base

    return output

convertToBase(200, 16)
```

```python
class TreeNode:
    def __init__(self, data=""):
        self.data = data
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None

    def add(self, value):
        tempNode = TreeNode(value)
        if(self.root == None):
            self.root = tempNode

        else:
            probe = self.root
            while(probe != None):
                if(probe.data >= value):
                    if(probe.left == None):
                        probe.left = tempNode
                        probe = None
                    else:
                        probe = probe.left

                else:
                    if(probe.right == None):
                        probe.right = tempNode
                        probe = None
                    else:
                        probe = probe.right

    def inOrderTraversal(self):
        return self.inOrder(self.root)

    def inOrder(self, node):
        if(node == None):
            return ""
        else:
            return self.inOrder(node.left) + node.data + self.inOrder(node.right)

    def postOrderTraversal(self):
        return self.postOrder(self.root)

    def postOrder(self, node):
        if(node == None):
            return ""
        else:
            return self.postOrder(node.left) + self.postOrder(node.right) + node.data
```

```python
# 1A -> 16^1 * 1 

def hex2dec(hexVal):
    hexString = "0123456789ABCDEF"
    returnVal = 0
    for i in range(len(hexVal)-1, -1, -1):
        character = hexVal[i]
        power = len(hexVal)-1-i

        returnVal += (16**(power)) * hexString.index(character)

    return returnVal

hex2dec("1A")

    

```
