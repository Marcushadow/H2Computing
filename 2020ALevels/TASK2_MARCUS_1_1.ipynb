```python
# Task 2.1


filename = "TESTDATA.txt"
sortedList = []

with open(filename) as fs:
    dataList = fs.read().split("\n")
    for value in dataList:
        sortedList.append(value.split(","))

sortedList = sortedList[:-1]
# print(sortedList)



def quicksort(arr, left = -1 , right = -1):
    if(left == right == -1):
        left = 0
        right = len(arr) - 1
    

    if(left < right):
        pivot = partition(arr, left, right)
        quicksort(arr, left, pivot - 1)
        quicksort(arr, pivot + 1, right)

def partition(arr, left, right):
    pivVal = arr[left][0]
    leftPtr = left + 1
    rightPtr = right

    while(leftPtr <= rightPtr):
        while((leftPtr <= rightPtr) and (arr[leftPtr][0] <= pivVal)):
            leftPtr += 1
        
        while((leftPtr <= rightPtr) and (arr[rightPtr][0] > pivVal)):
            rightPtr -= 1
        
        if(leftPtr <= rightPtr):
            arr[leftPtr], arr[rightPtr] = arr[rightPtr], arr[leftPtr]

    arr[rightPtr], arr[left] = arr[left], arr[rightPtr]
    return rightPtr

quicksort(sortedList)
uploadList = []

for value in sortedList:
    uploadList.append(",".join(value))

stringToUpload = "\n".join(uploadList)

with open("TESTDATA_Filename_order_Marcus_1_1.txt", "w") as fs:
    fs.write(stringToUpload)
    

```
