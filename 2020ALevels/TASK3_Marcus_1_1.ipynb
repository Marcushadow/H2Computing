```python
class Node:
    def __init__(self, value):
        self._value = value
        self._next = None

    def setValue(self, newValue):
        self._value = newValue
    
    def getValue(self):
        return self._value

    def setNext(self, newNext):
        self._next = newNext
    
    def getNext(self):
        return self._next

class DataStructure:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def setHead(self, newHead):
        self.head = newHead
    
    def getHead(self):
        return self.head

    def setTail(self, newTail):
        self.tail = newTail
    
    def getTail(self):
        return self.tail

    def isEmpty(self):
        return self.head == None
    
    def display(self):
        if not self.isEmpty():
            tempNode = self.head()
            while(tempNode != None):
                print(tempNode.getValue())
                tempNode = tempNode.getNext()
        else:
            print("Empty")



    
```

```python
# Task 3.2

class Stack(DataStructure):
    def __init__(self):
        super().__init__()

    def insert(self, value):
        tempNode = Node(value)
        if(self.isEmpty):
            self.head = tempNode
        
        else:
            tempNode.setNext(self.head)
            self.head = tempNode

    def delete(self):
        if(self.isEmpty):
            print("It is empty: ")

        else:
            print(self.head.getValue())
            self.head = self.head.getNext()
    
    def display(self):
        if not self.isEmpty():
            tempNode = self.head()
            while(tempNode != None):
                print(tempNode.getValue())
                tempNode = tempNode.getNext()
        else:
            print("Empty")
        
```
