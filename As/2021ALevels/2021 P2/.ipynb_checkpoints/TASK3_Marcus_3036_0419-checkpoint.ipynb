```python
# Task 3.1

class Node:
    def __init__(self, data = "", ptr = None):
        self.data = data
        self.ptr = ptr

class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self, integer_value):
        tempNode = Node(data = integer_value, ptr = self.head)
        self.head = tempNode
        
    def search(self, integer_value):
        probe = self.head
        found = False
        while(probe != None and not found):
            if(probe.data == integer_value):
                found = True
            probe = probe.ptr
            
        return found

    def delete(self, integer_value):
        if(not self.search(integer_value)):
            return None
        else:
            prev = None
            probe = self.head
            
            while(probe.data != integer_value):
                prev = probe
                probe = probe.ptr
                
            if(prev == None):
                self.head = probe.ptr
            else:
                prev.ptr = probe.ptr
        
    def count(self):
        probe = self.head
        cnt = 0
        while(probe != None):
            cnt += 1
            probe = probe.ptr
            
        return cnt
    
    def to_String(self):
        probe = self.head
        outputStr = "["
        while(probe != None):
            outputStr += str(probe.data)
            if(probe.ptr != None):
                outputStr += ", "
            probe = probe.ptr
        outputStr += "]"
        
        return outputStr
                
        
            
            
ll = LinkedList()
with open("Task3data.txt") as fs:
    data = fs.read().split("\n")
    for num in data:
        ll.insert(int(num))
        
    print(ll.to_String())
```

```python
# Task 3.2

class SortedLinkedList(LinkedList):
    def __init__(self):
        super().__init__()
        
    def insert(self, integer_value):
        tempNode = Node(data = integer_value)
        
        prev = None
        probe = self.head
        
        while(probe != None and probe.data < integer_value):
            prev = probe
            probe = probe.ptr
          
        tempNode.ptr = probe
        if(prev == None): 
            self.head = tempNode
        
        else:
            prev.ptr = tempNode
            
sll = SortedLinkedList()
with open("Task3data.txt") as fs:
    data = fs.read().split("\n")
    for num in data:
        sll.insert(int(num))
        
    print(sll.to_String())      
            
print(sll.search(94))
    
```

```python
# task 3.3

class Queue(LinkedList):
    def __init__(self):
        super().__init__()
        
    def enqueue(self, integer_value):
        prev = None
        probe = self.head
        
        while(probe != None):
            prev = probe
            probe = probe.ptr
        
        newNode = Node(data = integer_value)
        
        if(prev == None):
            self.head = newNode
            
        else:
            prev.ptr = newNode
    
    def dequeue(self):
#         self.count() is inherited
        if(self.count() == 0):
            return None
    
        else:
            value = self.head.data
            self.head = self.head.ptr
            return value

myQ = Queue()
with open("Task3data.txt") as fs:
    data = fs.read().split("\n")
    for num in data:
        myQ.enqueue(int(num))
            
            
for i in range(5):
    print(myQ.dequeue())
       
```
