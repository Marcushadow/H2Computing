```python
# Task 1.1
def enter_letter():
    letter = input("Enter a letter:")

    # check if user enters one single letter
    cond = letter.isalpha() and len(letter) == 1

    while not cond:
        print("Please enter a single letter in the range A-Z or a-z.")
        letter = input("Enter a letter:")
        cond = letter.isalpha() and len(letter) == 1
    return letter 

for i in range(3):
    print('Test Case:')
    letter = enter_letter()
    print('Letter:', letter)
    print()
   
```

```python
# Task 1.2


def enter_base():

    base = int(input("Enter a number base:"))
    # ensure base in correct range
    while base <= 10 or base >= 15:
        print("Please enter a base between 11 and 14 (inclusive).")
        base = input("Enter a number base:")
    return base

def convert(letter, base):
    # match the letters for base 
    base_match = '0123456789ABCD'

    # perform convertion
    denary = ord(letter)
    base_value = ''
    while denary != 0: # keep dividing until quotient becomes 0
        # collect remainders in reverse order
        base_value = base_match[denary % base] + base_value 
        denary //= base

    # display results
    print('{:<20} {}'.format('Letter', letter))
    print('{:<20} {}'.format('Denary', ord(letter)))
    print('{:<20} {}'.format('Number Base ' + str(base), base_value))

for i in range(2):
    letter = enter_letter()
    base = enter_base()
    convert(letter, base)
```

```python
# Task 2.1

# finding position to split
def split(name_list, low, high):
    pivot = low
    left = low
    right = high
    
    while left < right: 
        while left < right and name_list[left] < name_list[pivot]:
            left += 1
    
        while right > left and name_list[right] > name_list[pivot]:
            right -= 1
        
        # swap the two if they have not crossed over
        if left < right:
            temp = name_list[left]
            name_list[left] = name_list[right]
            name_list[right] = temp
    # update pivot
    name_list[pivot], name_list[right] = name_list[right], name_list[pivot]
    
    return right
    
# recurisve quicksort
def quicksort(name_list, low, high):
    if low < high:
        pos = split(name_list, low, high)
        quicksort(name_list, low, pos - 1)
        quicksort(name_list, pos + 1, high)
    
    
# generate list of filenames
with open("TESTDATA.txt") as f:
    name_list = []
    content_dict = {}
    for line in f:
        name, date, size = line.strip().split(',')
        name_list.append(name)
        
quicksort(name_list, 0, len(name_list) - 1)

print("List of filenames:")
for name in name_list:
    print(name)
    
with open("TESTDATA_Filename_order.txt",'w') as f:
    for name in name_list:
        f.write(name+'\n')
    # the file is auto-closed
        
        
```

```python
# Task 2.2

# bubble sort
def bubble(date_list):
    for i in range(len(date_list)):
        for j in range(len(date_list) - 1):
            if date_list[j] > date_list[j + 1]:
                temp = date_list[j]
                date_list[j] = date_list[j + 1]
                date_list[j + 1] = temp


def sort_display():
    # generate list of dates, dictionary with date as key 
    # and [name, size] as value
    with open("TESTDATA.txt") as f:
        date_list = []
        content_dict = {}
        for line in f:
            name, date, size = line.strip().split(',')
            
            if date not in date_list: # some dates are repeated
                date_list.append(date)
                
            if date in content_dict:
                content_dict[date].append([name, size])
            else:
                content_dict[date] = [[name, size]]
    
    bubble(date_list)
    
    print('{:<20} {:<20} {}'.format('Date Modified', 'File Name', 'Size in KB'))
    
    for date in date_list:
        for name, size in content_dict[date]:
            print('{:<20} {:<20} {}'.format(date, name, size))
              

# search with given name
def search_name():
    name_given = input("Enter filename:")
    with open("TESTDATA.txt") as f:
        name_list = []
        content_dict = {}
        for line in f:
            name, date, size = line.strip().split(',')
            name_list.append(name)
            content_dict[name] = size
    
    if name_given in name_list:
        print("File name:", name_given)
        print("File size:", content_dict[name_given], 'KB')
    else:
        print("File {} not found.".format(name_given))



# search with given date
def search_date():
    date_given = input("Enter modified date:")
    with open("TESTDATA.txt") as f:
        date_list = []
        content_dict = {}
        for line in f:
            name, date, size = line.strip().split(',')
            if date not in date_list:
                date_list.append(date)
            if date in content_dict:
                content_dict[date].append(name)
            else:
                content_dict[date] = [name]
            
    if date_given in date_list:
        print("File(s) modified on this date:", date_given)
        for name in content_dict[date_given]:
            print(name)
    else:
        print("No files modified on", date_given)


print('Test case 1: sort in date order')
sort_display()
print()

print('Test case 2: search for a file name that is contained in the file')
search_name()
print()

print('Test case 3: search for a file name that is not contained in the file')
search_name()
print()

print("Test case 4: display files modified on a given date")
search_date()
print()

print("Test case 5: display message showing that no files were modified on a given date")
search_date()
```

```python
# Task 3.1
class Node:
    def __init__(self, data, ptr):
        self.data = data
        self.ptr = ptr

class DataStructure:
    def __init__(self):
        self.Head = None
        self.Tail = None
        
    def display(self):
        if self.Head == None:
            print("Empty Linked List")
        else:
            cur = self.Head
            while cur != None:
                print(cur.data, end = ' ')
                cur = cur.ptr
        print()
```

```python
# Task 3.2        
class Stack(DataStructure):
    def insert(self,value):
        if self.Head == None: # when inserting into an empty stack, tail is udpated too
            self.Tail = Node(value, self.Head)
        self.Head = Node(value, self.Head) # insert value as head of stack
        
        
    def delete(self):
        if self.Head == None:
            print('Stack is empty')
        else:
            data = self.Head.data
            self.Head = self.Head.ptr # delete the head of stack
            if self.Head == None: # no more nodes after deletion, tail must be updated too
                self.Tail = None
            return data
        
    def display(self):
        if self.Head == None:
            print("Empty Stack!")
        else:
            print("Stack Contents:")
            cur = self.Head
            while cur != None:
                print(cur.data, end = ' ')
                cur = cur.ptr
            
        print()
```

```python
# Task 3.3
class Queue(DataStructure):
    def insert(self, value):
        if self.Head == None: # insert into empty queue
            self.Head = Node(value,self.Head)
            self.Tail = self.Head
        else: # insert into tail of queue
            self.Tail.ptr = Node(value, None)
            self.Tail = self.Tail.ptr
        
    def delete(self):
        if self.Head == None:
            print('Queue is empty')
        else:
            data = self.Head.data
            self.Head = self.Head.ptr # delete from head of queue
            if self.Head == None: # no more nodes after deletion, must update tail as well
                self.Tail = None
            return data
    
    def display(self):
        if self.Head == None:
            print("Empty Queue!")
        else:
            print("Queue Contents:")
            cur = self.Head
            while cur != None:
                print(cur.data, end = ' ')
                cur = cur.ptr
        print()
```
